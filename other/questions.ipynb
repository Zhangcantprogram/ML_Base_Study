2. 将数据转化为**字典格式**，再将**字典格式**转化为**数组格式**
3. 在特征工程的过程当中，为什么在训练集上需要进行**拟合并转换**，而在测试集上只需要进行**转换**？
    在训练集上**拟合（Fit）**是指让模型或处理器（例如 DictVectorizer）学习训练数据中的特征及其分布。
    **转换（Transform）**：在测试集上仅转换是因为我们希望使用在训练集上建立的映射关系。转换操作将应用已经学习到的特征结构，以便将测试数据映射到相同的特征空间。
    如果在测试集上重新拟合，会导致新的特征列被创建，这可能与训练集中的特征列不匹配，进而导致**特征数量不一致的问题**。
4. 特征数量的问题！！！！！！！
